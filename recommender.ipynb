### Genres with most occurences

```python
import ast

def extract_genres(genre_list):
    genre = []
    genre_list = ast.literal_eval(genre_list)
    for genre_dict in genre_list:
        genre.append(genre_dict["name"])

    return genre
```

```python
tmdb['new_genres'] = tmdb['genres'].apply(lambda x: extract_genres(x))
```

## Using Weighted average for each movie's Average Rating

```python
for i in range(1,10):
    print(tmdb['vote_count'].quantile(float('0.{}'.format(i))))
```

```python
# Calculate all the components based on the above formula
v=tmdb['vote_count']
R=tmdb['vote_average']
C=tmdb['vote_average'].mean()
m=tmdb['vote_count'].quantile(0.8)
```

```python
print(C,m)
```

```python
tmdb['weighted_average']=((R*v)+ (C*m))/(v+m)
```

```python
tmdb['weighted_average']
```

```python
tmdb.sort_values('weighted_average',ascending=False)[['original_title', 'vote_count', 'vote_average', 'weighted_average', 'popularity']].head(20)
```

```python
plt.figure(figsize=(13,8))
sns.barplot(x=tmdb.sort_values('weighted_average',ascending=False)['weighted_average'].head(10), y=tmdb.sort_values('weighted_average',ascending=False)['original_title'].head(10), data=tmdb.sort_values('weighted_average',ascending=False))
plt.xlim(4, 10)
plt.title('Best Movies by average votes', weight='bold')
plt.xlabel('Weighted Average Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
```

## Recommendation based on scaled weighted average and popularity score(Priority is given 50% to both)

```python
from sklearn.preprocessing import MinMaxScaler

scaling=MinMaxScaler()
scaled_tmdb=scaling.fit_transform(tmdb[['weighted_average','popularity']])
normalized_tmdb=pd.DataFrame(scaled_tmdb,columns=['weighted_average','popularity'])
normalized_tmdb.head()
```

```python
tmdb[['normalized_weight_average','normalized_popularity']]= normalized_tmdb
```

```python
tmdb.head(10)
```

```python
tmdb['score'] = tmdb['normalized_weight_average'] * 0.5 + tmdb['normalized_popularity'] * 0.5
```

```python
tmdb.sort_values(['score'], ascending=False)[['original_title', 'normalized_weight_average', 'normalized_popularity', 'score']].head(20)
```

```python
plt.figure(figsize=(16,6))

sns.barplot(x=tmdb.sort_values(['score'], ascending=False)['score'].head(10), y=tmdb.sort_values(['score'], ascending=False)['original_title'].head(10), data=tmdb.sort_values(['score'], ascending=False), palette='deep')

plt.title('Best Rated & Most Popular Blend', weight='bold')
plt.xlabel('Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
```
